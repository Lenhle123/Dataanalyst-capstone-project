#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("survey-data.csv")

# Step 1: Convert Age categories to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Step 2: Encode SOPartFreq into ordinal values
sopartfreq_map = {
    'I have never participated in Q&A on Stack Overflow': 0,
    'Less than once per month or monthly': 1,
    'A few times per month or weekly': 2,
    'A few times per week': 3,
    'Daily or almost daily': 4,
    'Multiple times per day': 5
}
df['SOPartFreqEncoded'] = df['SOPartFreq'].map(sopartfreq_map)

# Step 3: Clean data
df_clean = df.dropna(subset=['AgeNumeric', 'SOPartFreqEncoded', 'JobSat'])

# Step 4: Plot bubble chart
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_clean,
    x='AgeNumeric',
    y='SOPartFreqEncoded',
    size='JobSat',
    hue='JobSat',
    sizes=(20, 300),
    palette='coolwarm',
    alpha=0.6,
    legend=False
)

# Format the plot
plt.title('Bubble Plot of Age vs. Stack Overflow Participation Frequency\n(Bubble Size = Job Satisfaction)', fontsize=14)
plt.xlabel('Age (Midpoint of Age Group)')
plt.ylabel('Participation Frequency (Encoded)')
plt.yticks(
    ticks=list(sopartfreq_map.values()),
    labels=list(sopartfreq_map.keys()),
    fontsize=9
)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("survey-data.csv")

# Step 1: Convert age groups to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Step 2: Convert compensation and job satisfaction to numeric
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Step 3: Drop rows with missing or invalid data
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSat', 'AgeNumeric'])

# Optional: Filter out extremely high/low compensation
df_clean = df_clean[(df_clean['ConvertedCompYearly'] > 1000) & 
                    (df_clean['ConvertedCompYearly'] < 500000)]

# Step 4: Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df_clean,
    x='ConvertedCompYearly',
    y='JobSat',
    size='AgeNumeric',
    hue='AgeNumeric',
    sizes=(20, 300),
    palette='viridis',
    alpha=0.6,
    legend=False
)

# Formatting
plt.title('Bubble Plot of Compensation vs. Job Satisfaction\n(Bubble Size = Age)', fontsize=14)
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (0–10)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("survey-data.csv")

# Step 1: Map age categories to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Step 2: Drop missing Age or LanguageHaveWorkedWith
df_clean = df.dropna(subset=['AgeNumeric', 'LanguageHaveWorkedWith'])

# Step 3: Explode the LanguageHaveWorkedWith column into rows
df_clean['LanguageHaveWorkedWith'] = df_clean['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_clean.explode('LanguageHaveWorkedWith')

# Step 4: Count frequency of each language per age group
freq_df = (
    df_exploded.groupby(['LanguageHaveWorkedWith', 'AgeNumeric'])
    .size()
    .reset_index(name='Frequency')
)

# Step 5: Plot the bubble chart
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=freq_df,
    x='AgeNumeric',
    y='LanguageHaveWorkedWith',
    size='Frequency',
    hue='Frequency',
    sizes=(40, 600),
    palette='coolwarm',
    alpha=0.6,
    legend=False
)

# Final formatting
plt.title('Bubble Plot of Programming Language Popularity by Age Group')
plt.xlabel('Age (Midpoint of Age Group)')
plt.ylabel('Programming Language')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 1: Keep only relevant columns and drop rows with missing values
df_db = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()

# Step 2: Convert JobSat to numeric
df_db['JobSat'] = pd.to_numeric(df_db['JobSat'], errors='coerce')

# Step 3: Split semicolon-separated databases into lists
df_db['DatabaseWantToWorkWith'] = df_db['DatabaseWantToWorkWith'].str.split(';')

# Step 4: Explode the database list into separate rows
df_exploded = df_db.explode('DatabaseWantToWorkWith')

# Step 5: Group by database and calculate average JobSat and count
summary_df = (
    df_exploded
    .groupby('DatabaseWantToWorkWith')
    .agg(
        AvgJobSat=('JobSat', 'mean'),
        RespondentCount=('JobSat', 'count')
    )
    .reset_index()
)

# Optional: Filter to top 15 most popular databases for readability
summary_df = summary_df.sort_values(by='RespondentCount', ascending=False).head(15)

# Step 6: Plot the bubble chart
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=summary_df,
    x='AvgJobSat',
    y='DatabaseWantToWorkWith',
    size='RespondentCount',
    hue='RespondentCount',
    sizes=(100, 1000),
    palette='viridis',
    alpha=0.7,
    legend=False
)

# Final formatting
plt.title('Bubble Plot of Preferred Databases vs. Job Satisfaction\n(Bubble Size = Number of Respondents)', fontsize=14)
plt.xlabel('Average Job Satisfaction')
plt.ylabel('Preferred Database')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the data
df = pd.read_csv("survey-data.csv")

# Step 2: Keep necessary columns and clean
df = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()

# Convert numeric columns
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')
df['JobSat'] = pd.to_numeric(df['JobSat'], errors='coerce')

# Drop outliers and invalid entries
df = df[(df['ConvertedCompYearly'] > 1000) & (df['ConvertedCompYearly'] < 500000)]

# Step 3: Explode DevType (multiple roles per person separated by ;)
df['DevType'] = df['DevType'].str.split(';')
df_exploded = df.explode('DevType')

# Step 4: Group and summarize
summary_df = (
    df_exploded
    .groupby('DevType')
    .agg(
        AvgComp=('ConvertedCompYearly', 'mean'),
        AvgJobSat=('JobSat', 'mean'),
        Count=('JobSat', 'count')
    )
    .reset_index()
)

# Optional: Keep top 10 most common roles for readability
top_roles = summary_df.sort_values(by='Count', ascending=False).head(10)

# Step 5: Plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=top_roles,
    x='AvgComp',
    y='DevType',
    size='AvgJobSat',
    hue='AvgJobSat',
    sizes=(100, 800),
    palette='coolwarm',
    alpha=0.7,
    legend=False
)

# Final formatting
plt.title('Bubble Plot of Compensation Across Developer Roles\n(Bubble Size = Average Job Satisfaction)', fontsize=14)
plt.xlabel('Average Yearly Compensation (USD)')
plt.ylabel('Developer Role')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("survey-data.csv")

# Step 1: Map Age to numeric midpoints
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29.5,
    '35-44 years old': 39.5,
    '45-54 years old': 49.5,
    '55-64 years old': 59.5,
    '65 years or older': 70,
    'Prefer not to say': None
}
df['AgeNumeric'] = df['Age'].map(age_map)

# Step 2: Drop rows with missing Age or collaboration tools
df_clean = df.dropna(subset=['AgeNumeric', 'NEWCollabToolsHaveWorkedWith'])

# Step 3: Split and explode collaboration tools
df_clean['NEWCollabToolsHaveWorkedWith'] = df_clean['NEWCollabToolsHaveWorkedWith'].str.split(';')
df_exploded = df_clean.explode('NEWCollabToolsHaveWorkedWith')

# Step 4: Count frequency of each tool by age
freq_df = (
    df_exploded
    .groupby(['NEWCollabToolsHaveWorkedWith', 'AgeNumeric'])
    .size()
    .reset_index(name='Frequency')
)

# Optional: Keep top 10 most frequently used tools overall
top_tools = (
    freq_df.groupby('NEWCollabToolsHaveWorkedWith')['Frequency']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)
freq_df = freq_df[freq_df['NEWCollabToolsHaveWorkedWith'].isin(top_tools)]

# Step 5: Plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=freq_df,
    x='AgeNumeric',
    y='NEWCollabToolsHaveWorkedWith',
    size='Frequency',
    hue='Frequency',
    sizes=(50, 800),
    palette='viridis',
    alpha=0.7,
    legend=False
)

# Final formatting
plt.title('Bubble Plot of Collaboration Tools Used by Age Group', fontsize=14)
plt.xlabel('Age (Midpoint of Age Group)')
plt.ylabel('Collaboration Tool')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Step 1: Keep necessary columns and drop missing values
df_web = df[['WebframeWantToWorkWith', 'JobSat']].dropna()

# Convert JobSat to numeric
df_web['JobSat'] = pd.to_numeric(df_web['JobSat'], errors='coerce')

# Step 2: Split multiple web frameworks
df_web['WebframeWantToWorkWith'] = df_web['WebframeWantToWorkWith'].str.split(';')

# Step 3: Explode to separate rows
df_exploded = df_web.explode('WebframeWantToWorkWith')

# Step 4: Group by framework and compute stats
summary_df = (
    df_exploded.groupby('WebframeWantToWorkWith')
    .agg(
        AvgJobSat=('JobSat', 'mean'),
        RespondentCount=('JobSat', 'count')
    )
    .reset_index()
)

# Optional: Show top 15 most preferred frameworks
summary_df = summary_df.sort_values(by='RespondentCount', ascending=False).head(15)

# Step 5: Bubble plot
plt.figure(figsize=(14, 8))
sns.scatterplot(
    data=summary_df,
    x='AvgJobSat',
    y='WebframeWantToWorkWith',
    size='RespondentCount',
    hue='RespondentCount',
    sizes=(100, 800),
    palette='coolwarm',
    alpha=0.7,
    legend=False
)

# Final touches
plt.title('Bubble Plot of Preferred Web Frameworks vs. Job Satisfaction', fontsize=14)
plt.xlabel('Average Job Satisfaction')
plt.ylabel('Preferred Web Framework')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("survey-data.csv")

# Step 1: Drop rows with missing LanguageAdmired or Country
df_clean = df.dropna(subset=['LanguageAdmired', 'Country'])

# Step 2: Split the admired languages by semicolon
df_clean['LanguageAdmired'] = df_clean['LanguageAdmired'].str.split(';')

# Step 3: Explode to separate rows per admired language
df_exploded = df_clean.explode('LanguageAdmired')

# Step 4: Count frequency of admired tech by country
freq_df = (
    df_exploded.groupby(['Country', 'LanguageAdmired'])
    .size()
    .reset_index(name='Frequency')
)

# Optional: Keep top 10 countries by total responses
top_countries = (
    freq_df.groupby('Country')['Frequency']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)
freq_df = freq_df[freq_df['Country'].isin(top_countries)]

# Optional: Keep top 15 admired technologies overall
top_techs = (
    freq_df.groupby('LanguageAdmired')['Frequency']
    .sum()
    .sort_values(ascending=False)
    .head(15)
    .index
)
freq_df = freq_df[freq_df['LanguageAdmired'].isin(top_techs)]

# Step 5: Plot bubble chart
plt.figure(figsize=(16, 10))
sns.scatterplot(
    data=freq_df,
    x='Country',
    y='LanguageAdmired',
    size='Frequency',
    hue='Frequency',
    sizes=(50, 800),
    palette='coolwarm',
    alpha=0.7,
    legend=False
)

plt.title('Bubble Plot of Admired Technologies Across Countries', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Admired Technology')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
