### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python


# Step 2: Drop missing values in the compensation column
comp_data = df['ConvertedCompYearly'].dropna()

# Step 3: Plot distribution and histogram
plt.figure(figsize=(10, 6))
sns.histplot(comp_data, bins=50, kde=True, color='skyblue')
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter for full-time employed respondents
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Calculate median compensation
median_comp_full_time = full_time_df['ConvertedCompYearly'].median()

print(f"Median Yearly Compensation for Full-Time Employees: ${median_comp_full_time:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python


# Step 1: Filter necessary columns and drop missing values
df_filtered = df[['Country', 'ConvertedCompYearly']].dropna()

# Step 2: Filter countries with sufficient responses (e.g., >100)
country_counts = df_filtered['Country'].value_counts()
top_countries = country_counts[country_counts > 100].index
df_top_countries = df_filtered[df_filtered['Country'].isin(top_countries)]

# Optional: Remove extreme outliers using IQR method for better boxplot readability
Q1 = df_top_countries['ConvertedCompYearly'].quantile(0.25)
Q3 = df_top_countries['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df_top_countries[
    (df_top_countries['ConvertedCompYearly'] >= lower_bound) & 
    (df_top_countries['ConvertedCompYearly'] <= upper_bound)
]

# Step 3: Create the box plot
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_no_outliers, x='ConvertedCompYearly', y='Country', palette='viridis')
plt.title('Distribution of Yearly Compensation by Country (Top Responding Countries)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python


# Step 1: Drop rows with missing ConvertedCompYearly
df_clean = df.dropna(subset=['ConvertedCompYearly'])

# Step 2: Calculate IQR bounds for ConvertedCompYearly
Q1 = df_clean['ConvertedCompYearly'].quantile(0.25)
Q3 = df_clean['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 3: Filter rows within the IQR bounds
df_refined = df_clean[
    (df_clean['ConvertedCompYearly'] >= lower_bound) & 
    (df_clean['ConvertedCompYearly'] <= upper_bound)
]

# Step 4: Check dimensions
print(f"Original dataset size: {df.shape}")
print(f"After dropping NA (Comp only): {df_clean.shape}")
print(f"Refined dataset size (no comp outliers): {df_refined.shape}")
```
