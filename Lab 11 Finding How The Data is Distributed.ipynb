### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Display the column names
print("Column Names:")
print(df.columns.tolist())

# Display the data types of each column
print("\nData Types:")
print(df.dtypes)

# Display summary information about the dataframe
print("\nSummary Info:")
df.info()

# Optional: Display a statistical summary for numeric columns
print("\nStatistical Summary (Numeric Columns):")
print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python

# Display basic structure
print("Initial Data Shape:", df.shape)

# Step 1: Identify missing values
missing_values = df.isnull().sum()
print("\nMissing values per column:")
print(missing_values[missing_values > 0].sort_values(ascending=False))

# Step 2: Drop columns with more than 50% missing values
threshold = 0.5 * len(df)  # 50% of total rows
cols_to_drop = missing_values[missing_values > threshold].index
df.drop(columns=cols_to_drop, inplace=True)
print(f"\nDropped columns with >50% missing values: {list(cols_to_drop)}")
print("Data Shape after dropping columns:", df.shape)

# Step 3: Impute missing values in categorical columns with mode
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)

# Step 4: Impute missing values in numeric columns with median
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)

# Step 5: Confirm all missing values are handled
print("\nRemaining missing values per column (should be 0):")
print(df.isnull().sum().sum())

# Optional: Save cleaned dataset
# df.to_csv("survey-data-cleaned.csv", index=False)
```

- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

```python
## Write your code here
# Analyze Key Columns

# Step 1: Define the key columns to examine
key_columns = ['Employment', 'JobSat', 'YearsCodePro']

# Step 2: Calculate and display value counts for each column
for col in key_columns:
    if col in df.columns:
        print(f"\nValue Counts for '{col}':")
        print(df[col].value_counts(dropna=False))  # Includes missing if any were left
    else:
        print(f"\nColumn '{col}' not found in the dataset.")
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib_venn import venn2

# Step 1: Extract and split language responses into lists
def split_languages(column):
    return df[column].dropna().str.split(';').sum()

# Create Counter objects for each column
have_worked = Counter(split_languages('LanguageHaveWorkedWith'))
want_to_work = Counter(split_languages('LanguageWantToWorkWith'))

# Step 2: Convert to sets for comparison
langs_have = set(have_worked.keys())
langs_want = set(want_to_work.keys())

# Step 3: Venn Diagram
plt.figure(figsize=(8, 6))
venn2([langs_have, langs_want], set_labels=('Have Worked With', 'Want to Work With'))
plt.title('Programming Language Experience vs. Interest')
plt.show()

# Step 4: Grouped Bar Chart
# Combine counts for top languages (intersection only or top overall)
top_languages = list((have_worked + want_to_work).most_common(10))
langs = [lang for lang, _ in top_languages]
have_counts = [have_worked[lang] for lang in langs]
want_counts = [want_to_work[lang] for lang in langs]

# Plot grouped bar chart
x = range(len(langs))
plt.figure(figsize=(12, 6))
plt.bar(x, have_counts, width=0.4, label='Have Worked With', align='center')
plt.bar([i + 0.4 for i in x], want_counts, width=0.4, label='Want to Work With', align='center')
plt.xticks([i + 0.2 for i in x], langs, rotation=45)
plt.xlabel('Programming Languages')
plt.ylabel('Number of Respondents')
plt.title('Programming Languages: Experience vs. Interest')
plt.legend()
plt.tight_layout()
plt.show()


```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python


# Check actual column names to ensure they exist
print(df.columns.tolist())  # Inspect column names

# Use only countries with enough responses (e.g., >100) for better visibility
country_counts = df['Country'].value_counts()
top_countries = country_counts[country_counts > 100].index

# Filter data
filtered_df = df[df['Country'].isin(top_countries)]

# Create a crosstab: RemoteWork by Country
remote_by_country = pd.crosstab(filtered_df['Country'], filtered_df['RemoteWork'])

# --- Option 1: Grouped Bar Chart ---
remote_by_country_norm = remote_by_country.div(remote_by_country.sum(axis=1), axis=0)

remote_by_country_norm.plot(kind='bar', stacked=False, figsize=(14, 6), colormap='tab10')
plt.title('Distribution of Remote Work Status by Country')
plt.ylabel('Proportion of Respondents')
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work')
plt.tight_layout()
plt.show()

# --- Option 2: Heatmap ---
plt.figure(figsize=(14, 8))
sns.heatmap(remote_by_country_norm, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Heatmap: Remote Work Distribution by Country')
plt.xlabel('Remote Work Status')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
print(df['YearsCodePro'].value_counts())
print(df['JobSatPoints_1'].value_counts())
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Load dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Step 1: Clean 'YearsCodePro'
df['YearsCodePro_Clean'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})
df['YearsCodePro_Clean'] = pd.to_numeric(df['YearsCodePro_Clean'], errors='coerce')

# Step 2: Clean 'JobSat'
job_sat_mapping = {
    'Very satisfied': 5,
    'Slightly satisfied': 4,
    'Neither satisfied nor dissatisfied': 3,
    'Slightly dissatisfied': 2,
    'Very dissatisfied': 1
}
df['JobSat_Score'] = df['JobSatPoints_1'].map(job_sat_mapping)

# Step 3: Drop rows with missing values
corr_df = df[['JobSat_Score', 'YearsCodePro_Clean']].dropna()
print(f"Number of valid rows for correlation: {len(corr_df)}")

# Step 4: Proceed only if we have enough data
if len(corr_df) >= 2:
    pearson_corr, p_pearson = pearsonr(corr_df['JobSat_Score'], corr_df['YearsCodePro_Clean'])
    spearman_corr, p_spearman = spearmanr(corr_df['JobSat_Score'], corr_df['YearsCodePro_Clean'])

    # Step 5: Print results
    print(f"Pearson Correlation: {pearson_corr:.3f} (p = {p_pearson:.3f})")
    print(f"Spearman Correlation: {spearman_corr:.3f} (p = {p_spearman:.3f})")
else:
    print("Not enough valid data to compute correlation.")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
!pip install pandas

## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display actual column names
print(df.columns.tolist())

# Check if columns exist
if 'Employment' in df.columns and 'EdLevel' in df.columns:
    
    # Step 1: Create the cross-tab
    crosstab = pd.crosstab(df['EdLevel'], df['Employment'])
    
    # Step 2: Normalize by row (optional, for proportions)
    crosstab_percent = crosstab.div(crosstab.sum(axis=1), axis=0)

    # Step 3: Plot stacked bar chart
    crosstab_percent.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
    plt.title('Employment Status by Education Level')
    plt.xlabel('Education Level')
    plt.ylabel('Proportion of Respondents')
    plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
else:
    print("Required columns 'Employment' or 'EdLevel' not found in the dataset.")
```
